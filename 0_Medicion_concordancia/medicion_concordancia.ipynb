## 2. Análisis de la Concordancia.

```python
# Extraemos los valores de las columnas 'RESULTADO PCR' y 'DESC_RESULT_PRUEBA_RAPIDA' en arreglos de NumPy para
# poder realizar operaciones matriciales de manera más eficiente.
pcr = datos_consistentes["RESULTADO PCR"].values
res_pcr = pcr.reshape(1,-1)

prueba_rapida = datos_consistentes["DESC_RESULT_PRUEBA_RAPIDA"].values
res_p_r = prueba_rapida.reshape(1,-1)

print(f'\nForma de Res_PCR: {res_pcr.shape} \n\nRes_PCR:\n {res_pcr}')
print(f'\n\nForma de Res_prueba_rapida: {res_p_r.shape} \n\nRes_prueba_rápida:\n {res_p_r} \n')
```

* **Tabla general para el análisis de la concordancia**:

| PCR \ PR     | Positivo    | Negativo     | Total       |
| ------------ | :---------: | :----------: | :---------: |
| **Positivo** | a           | b            | _u_         |
| **Negativo** | c           | d            | _v_         |
| **Total**    | _r_         | _s_          | **N**       |

Donde:
* $u = a+b$
* $v = c+d$
* $r = a+c$
* $s = b+d$

```python
# Realizamos operaciones matriciales para obtener la matriz de confusión (en este caso, de concordancia).
q=(np.array([1,0]).reshape(2,1) == res_pcr).astype(np.int64)
l=(res_p_r.reshape(-1,1) == np.array([1, 0]).reshape(1,2)).astype(np.int64)

cm = np.matmul(q,l)
print(f'\nMatriz de concordancia: \n {cm} \n')
```

## 3. Obtención del porcentaje de concordancia.

La fórmula del **porcentaje de concordancia** es la siguiente:

$$
PC = \frac{a+d}{N} x 100
$$

```python
# Obtención del porcentaje de concordancia ocupando la fórmula.
porcentaje_concordancia = ((cm[0][0]+cm[1][1])/n)*100
print(f'\nEl porcentaje de concordancia es: {round(porcentaje_concordancia,3)}% \n')
```

---
## 4. Obtención del índice de Kappa.

La fórmula del **índice de Kappa ($\kappa$)** es la siguiente:

$$
\kappa = \frac{C_o-C_e}{1-C_e}
$$
Donde:
* $C_o$: proporción de concordancias observadas.
* $C_e$: concordancia debida al azar.

Sabemos que:
$$
C_o = \frac{a+d}{N}
$$

```python
# Cálculo de la proporción de concordancias observadas
c_o = (cm[0][0]+cm[1][1])/n
print(f'\nEl valor de la proporción de concordancias observadas Co es: {c_o} \n')
```

Por otro lado:
$$
C_e = \frac{a_e+d_e}{N}
$$

* Donde:

$$a_e = \frac{r*u}{N}$$ 

$$d_e = \frac{s*v}{N}$$

```python
# Cálculo de los valores esprados de las celdas 'a' y 'd' de la tabla de concordancia.
a_e = (r*u)/n
d_e = (s*v)/n
print(f'\nA_e = {a_e} \nD_e = {d_e} \n')
```

Si sustituímos los valores antes obtenidos, tenemos que:

```python
# Cálculo de la concordancia debida al azar.
c_e = (a_e+d_e)/n
print(f'\nEl valor de la concordancia debida al azar es: {c_e} \n')
```

Ya que tenemos todos los valores necesarios, podemos utilizar la fórmula de **$\kappa$**.

```python
# Cálculo del índice de Kappa.
k = (c_o-c_e)/(1-c_e)
print(f'\nEl índice de kappa es: {k} \n')
```

## 5. Obtención de los valores predictivos

Veamos las fórmulas que vamos a ocupar:

* **Valor Predictivo Positivo**
$$
VPP = \frac{VP}{VP+FP}
$$


* **Valor Predictivo Negativo**
$$
VPN = \frac{VN}{VN+FN}
$$

Volvamos a echar un ojo a la **tabla para el análisis de la concordancia**:

| PCR \ PR     | Positivo    | Negativo     |
| ------------ | :---------: | :----------: |
| **Positivo** | 33          | 86           |
| **Negativo** | 8           | 235          |

¿Cuáles son los verdaderos y falsos **positivos**? ¿Y de los **negativos**?

```python
# Obtención de los valores predictivos.
vp = cm[0][0]
fp = cm[1][0]
vn = cm[1][1]
fn = cm[0][1]

vpp = vp/(vp+fp)
vpn = vn/(fn+vn)
print(f'\nValor Predictivo Positivo: {round(vpp*100,3)}% \nValor Predictivo Negativo: {round(vpn*100,3)}% \n')
```
